```python
#  remove the unsuedd columns 
```

```python
df1=df.drop(['Crop_Year','District_Name'],axis = 1)
```

```python
# handling the categarical data by one hot coding method
```

```python
df1 = df.drop(['Crop_Year','District_Name'],axis = 1)
df2 = pd.get_dummies(df1)
df2.drop(columns='Crop_other oilseeds',axis=1)
df2.head()
```

```python
# now data is ready for learning algorithm
```

```python
# select the train_test data
```

```python
X=X=df2.drop('Production',axis =1)
y=df2[['Production']]
```

```python
# standardization of the data 
```

```python
from sklearn.preprocessing import StandardScaler

scalar=StandardScaler()

scaled=scalar.fit_transform(X)
```

```python
# split the data train and test
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(scaled,y,test_size=0.3,random_state=245)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# linearRegression
```

```python
from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
mlr.fit(X_train,y_train)
```

```python
#Printing the model coefficients
print('Intercept: ', mlr.intercept_)
print('Coefficient:', mlr.coef_)
```

```python
from sklearn.metrics import r2_score
```

```python
# prediction on test and train  data 
```

```python
y_test_pred = mlr.predict(X_test)
y_train_pred = mlr.predict(X_train)
a =  r2_score(y_train,y_train_pred)
b = r2_score(y_test,y_test_pred)
print("trainning = {} testing ={} ".format(a,b))
```

```python
# Random forest
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()

rf.fit(X_train, y_train)
```

```python
y_test_pred = rf.predict(X_test)
y_train_pred = rf.predict(X_train)
a =  r2_score(y_train,y_train_pred)
b = r2_score(y_test,y_test_pred)
print("trainning = {} testing ={} ".format(a,b))
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
print( mean_absolute_error(y_test,y_test_pred))
```

```python
# DecisionTree Regressor
```

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()

dt.fit(X_train,y_train)
```

```python
y_test_pred = dt.predict(X_test)
y_train_pred = dt.predict(X_train)
a =  r2_score(y_train,y_train_pred)
b = r2_score(y_test,y_test_pred)
print("trainning = {} testing ={} ".format(a,b))
```
